# Step 1 Reading data

```python
### Step 1 Reading data
df = pd.read_csv("../data/LendingClubLoan.csv")
df = df.drop('loan_status_Binary',axis=1)

# # Randomly select 5000 rows
# set.seed(33)
df = df.sample(n=5000,random_state=33, replace=False)

df.head()
```

```python
# statistics for quantitative variables
df.describe()
```

```python
# statistics for categorical variables
df.describe(include=[np.object])
```

```python
%matplotlib inline
# Plotting the density plot for loan_amnt distribution
sns.distplot(df['loan_amnt'], hist=False, color="b", kde_kws={"shade": True}, label='loan_amnt')
# sns.distplot(df['annual_inc'], hist=True, color="g",  label='DryBulbCelsius')
plt.xlabel('Money')
plt.ylabel('Density')
plt.title('Density Distribution of Temperatures')
plt.show()
```

The right-skewed nature of laon_amnt distribution implies people tend to borrow on higher side 

```python
# sns.pairplot(pdDf)
pdDf=df[['loan_amnt', 'int_rate', 'installment', 'annual_inc', 'home_ownership']]
pdDf.head()
plt.figure(figsize=(20,15))
sns.pairplot(pdDf, hue="home_ownership")
plt.show()
```

We see no co-relation between loan_mant and interest_rate, however there does exist an obvious co-relation between loan_amnt and #installments

# Step 2 Calculating distance

```python
gower_dist=pdist(df[['loan_amnt', 'int_rate', 'installment', 'annual_inc']], metric='euclidean')
print (gower_dist.shape)
gower_mat=np.matrix(gower_dist)
print (gower_mat.shape)
```

```python
gower_mat[gower_mat != min(gower_mat)].shape
```

```python
min(gower_mat[gower_mat != min(gower_mat)])
```

```python
np.argmin(gower_mat[gower_mat != min(gower_mat)])
```
