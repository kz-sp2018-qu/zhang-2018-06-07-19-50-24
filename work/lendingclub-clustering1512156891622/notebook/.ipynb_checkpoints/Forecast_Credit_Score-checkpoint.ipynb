## Data Preprocessing

```python
# selecting all laon before 2013
# Drop null values
columns=df.columns.values
df = df.dropna(axis=1, thresh=2)

df['year_issued'] = df.issue_d.apply(lambda x: int(x.split("-")[0]))
df_term = df[df.year_issued < 2013]
df_term.head()
```

## Creating custom features

```python
def convert(value):
    if value =='RENT':
        return 1
    if value == 'OWN':
        return 2
    if value == 'MORTGAGE':
        return 3
    
bad_indicators = [
    "Late (16-30 days)",
    "Late (31-120 days)",
    "Default",
    "Charged Off"
]

df_term['is_bad'] = df_term.loan_status.apply(lambda x: 1 if x in bad_indicators else 0)
df_term['home_ownership_num']=df_term.home_ownership.apply(convert)
features=['fico_range_low','fico_range_high','inq_last_6mths','home_ownership_num','annual_inc','loan_amnt']
is_bad = df_term['is_bad']
df_term = df_term[features]

# Normalizing df_term
means = df_term.mean(axis=0)
stds = df_term.mean(axis=0)
# df_term = (df_term - df_term.mean(axis=0))/df_term.std(axis=0)
df_term['is_bad']=is_bad
df_term.head()
```

# Predictive Analysis

##  1) Seperating data into training and testing dataset

```python
# Sample train and test data : we are using 80% of the dataset as training data
df_term = df_term.fillna(method='backfill')
predictors = df_term.loc[:, df_term.columns!='is_bad']
classes = df_term.loc[:,df_term.columns=='is_bad']

training_set_size = int(len(df_term) * 0.8)
test_set_size = len(df_term) - training_set_size

training_predictors = predictors[:training_set_size]
training_classes = classes[:training_set_size]
test_predictors = predictors[training_set_size:]
test_classes = classes[training_set_size:]

training_predictors.describe()
# training_classes.head()
```

## 2) Defining functions to perform analysis

```python
# Use passed function to build the model and train on the dataset
def buildModel(function):
    model=function.fit(training_predictors.as_matrix(), training_classes.as_matrix()[:,0])
    return model

# Predict on the test data using model created above
def makePrediction(function):
    predicted=function.predict(test_predictors.as_matrix())
    return predicted

# Calculate the accuracy achieved
def calcAccuracy(function,predicted):
#     temp=predicted==test_classes
#     print (temp)
#     accuracy=len(temp[temp==True])/len(temp)
#     print ("Accuracy using Logistic Regression Model = ",accuracy,"\n")

#     Or you can compute accuracy using below function
    accuracy = function.score(test_predictors.as_matrix(),test_classes.as_matrix()[:,0])
    print ("Accuracy using Logistic Regression Model = ",accuracy )

    # Generating the classfication Report
    print ("Classification Report : \n\n",
           metrics.classification_report(test_classes.as_matrix()[:,0], predicted))
    
    # Generating confusion matrix
    print ("Confusion Matrix : \n",
           metrics.confusion_matrix(test_classes.as_matrix()[:,0], predicted))
            

    # Verifying model using 10 fold cross validation
    scores1 = cross_val_score(estimator=function, X=training_predictors.as_matrix(), 
                         y=training_classes.as_matrix()[:,0], scoring='accuracy')
    print ("\nMean accuracy validated using Cross Validation: ",scores1.mean())
    
    return metrics.classification_report(test_classes.as_matrix()[:,0], predicted) , accuracy

# Calculate credit score
def calculate_score(log_odds):
    # 300 baseline + (40 points equals double risk) * odds
    return 300 + (40 / np.log(2)) * (-log_odds)

def cal_credit(model):
    probs = model.predict_proba(test_predictors.as_matrix())[:,1]
    log_probs = model.predict_log_proba(test_predictors.as_matrix())[:,1]
    scores = calculate_score(log_probs)
    return scores, probs

# Calling above functions sequentially
def runAnalysis(function, model_name):
    model=buildModel(function)    
#     joblib.dump({model_name: model}, "../output/"+model_name + '.pkl',
#                     compress=9)
    with open("../output/"+model_name+'.pkl', 'wb') as handle:
        pickle.dump({model_name:model}, handle, protocol=pickle.HIGHEST_PROTOCOL)
    predicted=makePrediction(function)
    classificationReport, acc=calcAccuracy(function,predicted)
    credit_scores, probs = cal_credit(model)
    print ("len of probs ",len(probs))
    with open("../output/defaultProbs.pkl", 'wb') as handle:
        pickle.dump(probs, handle, protocol=pickle.HIGHEST_PROTOCOL)
        
    with open("../output/probsjson", 'w') as outfile:
        json.dump({"values":list(probs)}, outfile)
    
    return probs
```

## 4) Predict is_bad using Logistic Regression

```python
# Predicting the is_bad (whether the applicant is likely to default or not) using Logistic Regression
logistic = LogisticRegression()
logistic_creditScore=runAnalysis(logistic, 'LogisticRegression')
```

## 4) Predict is_bad trend using RandomForestClassifier

```python
# Predicting the is_bad (whether the applicant is likely to default or not) using RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)
rf_creditScore=runAnalysis(clf, 'RandomForestClassifier')
```

# 5) Predict is_bad trend using Neural Network MLPClassifier

```python
# Predicting the is_bad (whether the applicant is likely to default or not) using Neural Network MLPClassifier
clf = MLPClassifier(solver='lbfgs', alpha=1e-2,
                     hidden_layer_sizes=(2, ), random_state=1,activation='logistic',max_iter=1000)
nn_creditScore=runAnalysis(clf,'NeuralNetwork')
```
